```python
import math

%matplotlib inline

def fun_cos(x,n):
  approx_cos = 0
  for i in range(n):
    coef = (-1)**i
    num = x**(2*i)
    denum = math.factorial(2*i)
    approx_cos += (coef*num)/denum
  return approx_cos
```

```python
import numpy as np
import matplotlib.pyplot as plt

angles = np.arange(-2*np.pi,2*np.pi,0.1)
p_cos = np.cos(angles)

fig, ax = plt.subplots()
ax.plot(angles,p_cos)

my_list = [-7,7]

for i in range(1,6):
  t_cos = [fun_cos(angle,i) for angle in angles]
  ax.plot(angles,t_cos)

ax.set_ylim(-4,4)
plt.show()
```

```python
import math
import matplotlib.pyplot as plt
import numpy as np

def func_sin(x,n):
  approx_sin = 0
  for i in range(n):
    coef = (-1)**i
    num = x**(2*i+1)
    denum = math.factorial(2*i+1)
    approx_sin += (coef*num)/denum
  return approx_sin

angles = np.arange(-2*np.pi,2*np.pi,0.1)
fig, ax = plt.subplots()

p_sin = np.sin(angles)
ax.plot(angles,p_sin)
t_sin = [func_sin(angle,2) for angle in angles] #3th order approximation
ax.plot(angles,t_sin)
ax.set_ylim(-8,8)

plt.legend(['sin function','Taylor series 2th terms'])
plt.show()
```

```python
#Bisection Method
import math
import numpy as np
import matplotlib.pyplot as plt

def func_f1(c):
  f1 = (667.38/c) * (1 - np.exp(-0.146843*c)) - 40
  return f1

xx = np.linspace(0,18,300)
yy = [func_f1(y) for y in xx]

fig = plt.figure(figsize=(6,6))
plt.plot(xx,yy)
plt.show()

iter = 2

for i in range(1,10):
  print("{} -> func_f({}): {}" .format(i,iter,func_f1(iter)))
  iter = iter + 2


xl = 14
xu = 16
xr = xu
imax = 100
es = 0.005
iter = 0

while True:
  xrold = xr
  xr = (xl + xu) / 2
  iter = iter + 1

  if xr != 0:
    ea = abs((xr - xrold) / xr) * 100

  test = func_f1(xl) * func_f1(xr)

  if test < 0:
    xu = xr
  elif test > 0:
    xl = xr
  else:
    ea = 0

  print("iter[{}], xl[{}], xu[{}], xr[{}], ea[{}], es[{}]" .format(iter, xl, xu, xr, ea, es))

  if ea < es or iter >= imax:
    break

Bisect = xr

print(">>> Bisect:",Bisect)
```

```python
#Standard False Position Method
import matplotlib.pyplot as plt
import math
import numpy as np

def func_f(x):
  f1 = x**10 - 1
  return f1

xx = np.linspace(0,2,500)
yy = [func_f(y) for y in xx]

fig = plt.figure(figsize = (2,10))
ax = fig.add_subplot(1,1,1)

plt.plot(xx,yy)
plt.show()

xl = 0
xu = 1.3
iter = 0
imax = 100
es = 0.005

while True:
  xrold = xr
  xr = xu - (func_f(xu) * (xl - xu) / (func_f(xl) - func_f(xu)))
  iter = iter + 1
  if xr != 0:
    ea = abs((xr - xrold) / xr) * 100
  test = func_f(xl) * func_f(xr)
  if test < 0 :
    xu = xr
  elif test > 0 :
    xl = xr
  else:
    ea = 0
  print("iter[{}] xl[{}], xu[{}], xr[{}], ea[{}], es[{}]" .format(iter,xl,xu,xr,ea,es))
  if ea < es or iter >= imax :
    break

FalsePos = xr

print("False Position f({}) = {}" .format(FalsePos,func_f(FalsePos)))
```

```python
#Modified False Position Method
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

def func_f1(x):
  f1 = x**10 - 1
  return f1

xx = np.linspace(0,2,500)
yy = [func_f1(y) for y in xx]

fig = plt.figure(figsize = (3,10))
ax = fig.add_subplot(1,1,1)

plt.plot(xx,yy)
plt.show()

xl = 0
xu = 1.3
iter = 0
imax = 100
es = 0.005
fl = func_f1(xl)
fu = func_f1(xu)

while True:
  xrold = xr
  xr = xu - ((fu * (xl - xu)) / (fl - fu))
  fr = func_f1(xr)
  iter = iter + 1

  if xr != 0 :
    ea = abs((xr - xrold) / xr) * 100
  test = fl * fr
  
  if test < 0 :
    xu = xr
    fu = func_f1(xu)
    iu = 0
    il = il + 1
    if il >= 2 :
      fl = fl/2
  elif test > 0 :
      xl = xr
      fl = func_f1(xl)
      il = 0
      iu = iu + 1
      if iu >= 2 :
        fu = fu / 2
  else:
      ea = 0

  print("iter[{}], xl[{}], xu[{}], xr[{}], ea[{}], es[{}]" .format(iter,xl,xu,xr,ea,es))
  
  if ea < es or iter >= imax :
    break

ModFalsePos = xr
print(">>>Modified Version of False Position Method: f({}) = {}".format(ModFalsePos, func_f1(ModFalsePos)))
```

```python
#Fixed Point Iteration Method
# f(x) = e^-x - x
import numpy as np
import math
plt.style.use('seaborn-whitegrid')
def func_h(x):
  f = np.exp(-x) - x
  return f

#x=g(x)=e^-x
def func_g(x):
  g = np.exp(-x)
  return g

xx = np.linspace(0,2,300)
yy = [func_h(y) for y in xx]

fig = plt.figure(figsize = (5,5))

plt.plot(xx,yy)
plt.show()

#anticipated value
x0 = 0
#initializing
iter = 0
es = 0.005
imax = 100
xr = x0

while True:
  xrold = xr
  xr = func_g(xrold)
  iter = iter + 1
  if xr != 0 :
    ea = abs((xr - xrold) / xr) * 100
  print("iter[{}] xr[{}] ea[{}] es[{}] " .format(iter,xr,ea,es))
  if ea < es or iter >= imax:
    break

Fixpt = xr
print("Fixed Point Iteration Method: {}" .format(Fixpt))
print("g({}) = {}" .format(Fixpt,func_g(Fixpt)))
```

```python
#Newton Raphson Method
# f(x) = e^-x - x
import numpy as np
import math

#Representation of f(x)
def func_h(x):
  f = np.exp(-x) - x
  return f

#f'(x)
def func_dev(x):
  dev = -np.exp(-x) - 1
  return dev

#xi+1 = xi - (f(xi) / f'(xi))
def func_fin(x):
  fin = x - (func_h(x) / func_dev(x))
  return fin

#anticipated value
x0 = 0
#initialization
xr = x0
iter = 0
imax = 100
es = 0.005

while True:
  xrold = xr
  xr = func_fin(xrold)
  iter = iter + 1
  if xr != 0 :
    ea = abs((xr - xrold) / xr) * 100
  print("iter[{}] xr[{}] ea[{}] es[{}]" .format(iter,xr,ea,es))
  if ea < es or iter >= imax :
    break

NewtonRaphson = xr
print(">>> Newton-Raphson: ",NewtonRaphson)
```

```python
#Secant Method
# f(x) = e^-x - x
import numpy as np
import math

delta = 0.1

def func_h(x):
  f = np.exp(-x) - x
  return f

def func_g(x):
  g = x - (x+(x+delta)-x)*(func_h(x))/(func_h(x+(x+delta)) - func_h(x))
  return g

iter = 0
imax = 100
x0 = 1
es = 0.0000005
xr = x0

while True:
  xrold = xr
  xr = func_g(xrold)
  iter = iter + 1 
  if xr != 0 :
    ea = abs((xr - xrold) / xr) * 100
  print("iter[{}] xr[{}] ea[{}] es[{}] " .format(iter, xr, ea, es))
  if ea < es or iter >= imax :
    break
Secant = xr
print(">>> Secant Method: {}" .format(Secant))
```

```python

#Muller Method
# f(x) = x^3 - 13x -12
#x0=4.5, x1=5.5, x2=5
import numpy as np
import math
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
def func_h(x):
  f = (x**3) - (13 * x) -12
  return f

xx = np.linspace(0,6,200)
yy = [func_h(y) for y in xx]

fig = plt.figure(figsize = (5,2))

plt.plot(xx,yy)
plt.show()

xr = 5
h = 0.1

x2 = xr
x1 = xr + (h*xr)
x0 = xr - (h*xr)
maxit = 100
eps = 1e-11
iter = 0

print("x0:{} , x1:{}  , x2:{}" .format(x0,x1,x2))

while True:
  xrold = xr
  iter = iter + 1
  h0 = x1 - x0
  h1 = x2 - x1
  d0 = (func_h(x1) - func_h(x0)) / h0
  d1 = (func_h(x2) - func_h(x1)) / h1
  a = (d1 - d0) / (h1 + h0)
  b = (a * h1) + d1
  c = func_h(x2)
  rad = math.sqrt(b*b - 4*a*c)

  if abs(b + rad) > abs(b - rad):
    den = b + rad
  else:
    den = b - rad
  dxr = (-2 * c) / den
  xr = x2 + dxr
  print("iter[{}] , xr[{}] , dxr[{}] " .format(iter, xr, dxr))

  if(abs(dxr) < eps * xr or iter >= maxit):
    break
  x0 = x1
  x1 = x2
  x2 = xr

print(">>> Muller Method: {}" .format(xr))
```

```python
#Finite Divided Difference approximation of Derivatives
#f(x) = -0.1x^4 - 0.15x^3 - 0.5x^2 -0.25x +1.25

import math
import numpy as np

def origin_func(x):
  f = -0.1 * (x**4) - 0.15 * (x**3) - 0.5 * (x**2) - 0.25 * x + 1.25
  return f

def dev_func(x):
  ff = -0.4 * (x**3) - 0.45 * (x**2) -(1.0 * x) - 0.25
  return ff

h = 0.5
xa = 0 #xi-1
xb = 0.5 #xi
xc = 1.0 #xi+1

fa = origin_func(xa)
fb = origin_func(xb)
fc = origin_func(xc)

exact_val = dev_func(h)

#Backward divided difference
a_diff = (fb - fa) / (xb - xa)
a_et = abs((exact_val - a_diff) / exact_val) * 100

#Forward divided difference
b_diff = (fc - fb) / (xc - xb)
b_et = abs((exact_val - b_diff) / exact_val) * 100

#Centered divided difference
c_diff = (fc - fa) / (xc - xa)
c_et = abs((exact_val - c_diff) / exact_val) * 100

print(">>> Backward: {}  et:{} " .format(a_diff, a_et))
print(">>> Forward: {}  et:{} " .format(b_diff, b_et))
print(">>> Centered: {}  et:{} " .format(c_diff, c_et))
print(">>> Exact Value: {} " .format(exact_val))
```

```python
#step size에 따른 finite difference(Centered Method)
import math
import numpy as np

def origin_func(x):
  f = -0.1 * (x**4) - 0.15 * (x**3) - 0.5 * (x**2) - 0.25 * x + 1.25
  return f

def dev_func(x):
  ff = -0.4 * (x**3) - 0.45 * (x**2) -(1.0 * x) - 0.25
  return ff

h = 1.
x = float(input("x:"))
n = int(input("n:"))

xa = 0
xb = 0

#value of f'(x)
exac_val = dev_func(x)
print("step size          finite difference         true error")
for i in range(0,n):
  xa = (x-h)
  xb = (x+h)
  diff = (origin_func(xb) - origin_func(xa)) / (2 * h)
  true_error = abs((exac_val - diff))
  h = h/10
  i = i + 1
  print("{}, {}, {}  " .format(h,diff,true_error))
```
