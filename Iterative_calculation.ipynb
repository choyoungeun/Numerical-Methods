```python
x = int(input("x 숫자를 입력하세요 : ")) 
es = float(input("es 숫자를 입력하세요 : ")) 
maxit = int(input("MaxIter 숫자를 입력하세요 : ")) 

iter = 1
sol = 1
ea = 100
fac = 1
# iterative calculation
while True:
  solold = sol
  fac = fac * iter
  sol = sol + x ** iter / fac
  iter = iter + 1
  if sol != 0 : ea = abs((sol - solold) / sol) * 100
  print("iter=", iter, "  ea=", ea, "  es=", es, "  sol=", sol)
  if ea <= es or iter >= maxit : break
```

```python
import math;

x = int(input("e의 지수값을 입력하시오 : "));

sum=1.;
term = 1.;
test=0.;
i=0;

while True :
  if sum == test : break;
  i=i+1;
  test=sum;  
  term = (x ** i) / math.factorial(i);  
  sum = sum + term;
  print("i=", i, "  term=", term, "  sum=", sum);
  
print("exact value = ", math.exp(x));
```
